```python
%%writefile p1.py
# %load p1.py
#Import librariers
import pandas as pd
import csv, math
import numpy as np
from IPython.display import display, HTML


#Schools file
file_name_school = "Resources/schools_complete.csv"
schools_data = pd.read_csv(file_name_school)


#student  file
file_path_student = "Resources/students_complete.csv"
students_data = pd.read_csv(file_path_student)


# School and Student file 
school_data_complete = pd.merge(students_data, schools_data, how="left", on=["school_name", "school_name"])
#print(school_data_complete.head())


#Calculate the average math score 
math_mean=school_data_complete["math_score"].mean()
#Format average math score 
math_mean= "{:.2f}".format(math.ceil(math_mean))
#print(math_mean)



#Calculate the average reading score 
reading_mean=school_data_complete["reading_score"].mean()
#Format average reading score 
reading_mean= "{:.2f}".format(math.ceil(reading_mean))
#print(reading_mean)


#Calculate the total budget
total_budget=school_data_complete["budget"].unique().sum()
#Format  budget
total_budget= "{:,.2f}".format(total_budget)
#print(total_budget)


#Calculate the total number of students
total_students=school_data_complete["size"].unique().sum()


#Calculate the total number of Schools
total_schools=len(school_data_complete["school_name"].unique())
s_name= school_data_complete["school_name"].unique()


#Calculate the percentage of students with a passing math score (70 or greater)
df1=school_data_complete[school_data_complete['math_score'] >=70].count()
ave_math=df1['math_score']/school_data_complete['math_score'].count() * 100

#Calculate the percentage of students with a passing Reading score (70 or greater)
df2=school_data_complete[school_data_complete['reading_score'] >=70].count()
ave_reading=df2['reading_score']/school_data_complete['math_score'].count() * 100
#print("{:.2f}%".format(math.ceil(ave_reading)))

 


ave_reading=df2['reading_score']/school_data_complete['math_score'].count() * 100

overall_pass = school_data_complete[(school_data_complete['math_score'] >= 70) & (school_data_complete['reading_score'] >= 70)]['school_name'].count()/total_students *100
#print(overall_pass)
#average of  joint population of math and reading is not like adding separate average and adding together and divide  by 2
#overall_pass = (ave_math + ave_reading) / 2  could be wrong
#print(overall_pass)
district_summary = pd.DataFrame({
    "Total Schools": [total_schools],
    "Total Students": [total_students],
    "Total Budget": [total_budget],
    "Average Reading Score": [reading_mean],
    "Average Math Score": [ave_math],
    "% Passing Reading":[ave_reading],
    "% Passing Math": [ave_math],
    "Overall Passing Rate": [overall_pass]
})
#organize     Columns per documnet(notebook)
ds1=  district_summary[["Total Schools", 
                        "Total Students",  
                        "Total Budget", 
                        "Average Reading Score",
                        "Average Math Score",
                        "% Passing Reading",
                        "% Passing Math",
                        "Overall Passing Rate"]]


display(ds1)
#ghassan
#per student budget
per_stu_budget = schools_data.set_index('school_name')['budget']/schools_data.set_index('school_name')['size']

group_by_school_name = school_data_complete.set_index('school_name').groupby(['school_name'])
avg_math_per_school = group_by_school_name['math_score'].mean()
avg_reading_per_school = group_by_school_name['reading_score'].mean()
school_types = schools_data.set_index('school_name')['type']
stu_per_sch = group_by_school_name['Student ID'].count()
pass_math = school_data_complete[school_data_complete['math_score'] >= 70].groupby('school_name')['Student ID'].count()/stu_per_sch *100 
pass_read = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('school_name')['Student ID'].count()/stu_per_sch  *100
overall = school_data_complete[(school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >= 70)].groupby('school_name')['Student ID'].count()/stu_per_sch * 100 

##
## school budget
sch_budget = schools_data.set_index('school_name')['budget']

students_summary=pd.DataFrame({"School Type":school_types,
    "Total Students":stu_per_sch,
    "Total School Budget":sch_budget,
    "Per Student Budget":per_stu_budget,
    "Average Math Score":avg_math_per_school,
    "Average Reading Score":avg_reading_per_school,
    "% Passing Math":pass_math,
    "% Passing Reading":pass_read,
    "Overall Passing Rate":overall})


ds2= students_summary[['School Type', 
                          'Total Students', 
                          'Total School Budget', 
                          'Per Student Budget', 
                          'Average Math Score', 
                          'Average Reading Score',
                          '% Passing Math',
                          '% Passing Reading',
                          'Overall Passing Rate']]
display(ds2)

sorty=ds2.sort_values("Overall Passing Rate",ascending=False)
display(sorty.head(5))

sorty1=ds2.sort_values("Overall Passing Rate")
sorty2=sorty.tail()
reversed_df = sorty2.iloc[::-1]
display(reversed_df.head())


#Create a table that lists the average Math Score for students of each grade level (9th, 10th, 11th, 12th) at each school.
ninth_math = students_data[students_data['grade'] == '9th'].groupby('school_name')["math_score"].mean()
n10th_math = students_data[students_data['grade'] == '10th'].groupby('school_name')["math_score"].mean()
n11th_math = students_data[students_data['grade'] == '11th'].groupby('school_name')["math_score"].mean()
n12th_math = students_data[students_data['grade'] == '12th'].groupby('school_name')["math_score"].mean()
dfmath=pd.concat([ninth_math,n10th_math,n11th_math,n12th_math],axis=1)
dfmath.columns = ['9th', '10th', '11th','12th']
display(dfmath)
#Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.
ninth_reading = students_data[students_data['grade'] == '9th'].groupby('school_name')["reading_score"].mean()
n10th_reading = students_data[students_data['grade'] == '10th'].groupby('school_name')["reading_score"].mean()
n11th_reading = students_data[students_data['grade'] == '11th'].groupby('school_name')["reading_score"].mean()
n12th_reading = students_data[students_data['grade'] == '12th'].groupby('school_name')["reading_score"].mean()
dfread=pd.concat([ninth_reading,n10th_reading,n11th_reading,n12th_reading],axis=1)
dfread.columns = ['9th', '10th', '11th','12th']
display(dfread)
'''
Summary of Schools by Score with  Spending, Size, and Type
--------------------------------------------------------
'''
'''
    Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
        Average Math Score
        Average Reading Score
        % Passing Math
        % Passing Reading
        Overall Passing Rate (Average of the above two)
'''
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
school_data_complete['spending_bins'] = pd.cut(school_data_complete['budget']/school_data_complete['size'], spending_bins, labels=group_names)
school_data_complete_group=school_data_complete.groupby('spending_bins') 
avg_math = school_data_complete_group['math_score'].mean()
#print(avg_math)
avg_read = school_data_complete_group['reading_score'].mean()
#print(avg_read)
pass_math = school_data_complete[school_data_complete['math_score'] >= 70].groupby('spending_bins')['Student ID'].count()/school_data_complete_group['Student ID'].count() *100
#print(pass_math)
pass_read = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('spending_bins')['Student ID'].count()/school_data_complete_group['Student ID'].count() *100
#print(pass_read)
overall = school_data_complete[(school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >= 70)].groupby('spending_bins')['Student ID'].count()/school_data_complete_group['Student ID'].count()  * 100

summary_by_spending=pd.DataFrame({"Average Math Score":avg_math,
        "Average Reading Score":avg_read,
        "% Passing Math":pass_math,
        "% Passing Reading":pass_read,
        "Overall Passing Rate":overall})
#organize columns
summary_by_spending = summary_by_spending[[
    "Average Math Score",
    "Average Reading Score",
    '% Passing Math',
    '% Passing Reading',
    "Overall Passing Rate"
]]
display(summary_by_spending)

'''
Scores by School Size
-------------------------
'''
size_bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
school_data_complete['size_bins'] = pd.cut(school_data_complete['size'], size_bins, labels=group_names)

school_data_complete_group=school_data_complete.groupby('size_bins')

avg_math = school_data_complete_group['math_score'].mean()
#print(avg_math)
avg_read = school_data_complete_group['reading_score'].mean()
#print(avg_read)
pass_math = school_data_complete[school_data_complete['math_score'] >= 70].groupby('size_bins')['Student ID'].count()/school_data_complete_group['Student ID'].count()
#print(pass_math)
pass_read = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('size_bins')['Student ID'].count()/school_data_complete_group['Student ID'].count()
#print(pass_read)
overall = school_data_complete[(school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >= 70)].groupby('size_bins')['Student ID'].count()/school_data_complete_group['Student ID'].count()


summary_by_size=pd.DataFrame({"Average Math Score":avg_math,
        "Average Reading Score":avg_read,
        "% Passing Math":pass_math,
        "% Passing Reading":pass_read,
        "Overall Passing Rate":overall})

#organize columns
summary_by_size = summary_by_size[[
    "Average Math Score",
    "Average Reading Score",
    '% Passing Math',
    '% Passing Reading',
    "Overall Passing Rate"
]]
display(summary_by_size)
'''
Scores by School Type
-------------------------
'''
group_by_type = school_data_complete.groupby("type")

# find avg, percent  from grouped df 
avg_math = group_by_type['math_score'].mean()
avg_read = group_by_type['math_score'].mean()
pass_math = school_data_complete[school_data_complete['math_score'] >= 70].groupby('type')['Student ID'].count()/group_by_type['Student ID'].count()* 100
pass_read = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('type')['Student ID'].count()/group_by_type['Student ID'].count()* 100
overall = school_data_complete[(school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >= 70)].groupby('type')['Student ID'].count()/group_by_type['Student ID'].count() * 100

# dataframe of summary by type        
summary_by_type = pd.DataFrame({
    "Average Math Score": avg_math,
    "Average Reading Score": avg_read,
    '% Passing Math': pass_math,
    '% Passing Reading':pass_read,
    "Overall Passing Rate": overall})
    
#organize columns
summary_by_type = summary_by_type[[
    "Average Math Score",
    "Average Reading Score",
    '% Passing Math',
    '% Passing Reading',
    "Overall Passing Rate"]]
display(summary_by_type)
```
